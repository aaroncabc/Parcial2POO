camión.

la localización geográfica y tiempo en el que pasó en su recorrido.

resultado de la separación se almacena en el sistema asociado al turno: cuántas toneladas

```python
from datetime import datetime

class TrashCity:
    def __init__(self):
        self.empleados = []
        self.camiones = []
        self.turnos = []

    def contratarEmpleado(self, empleado):
        self.empleados.append(empleado)

    def despedirEmpleado(self, empleado):
        self.empleados.remove(empleado)

    def agregarCamion(self, camion):
        self.camiones.append(camion)

    def eliminarCamion(self, camion):
        self.camiones.remove(camion)

    def cargaDiaN(self, dia):
        for turno in self.turnos:
            if turno.inicio.date() == dia.date():
                print(f"Carga del turno {turno.id} del día {dia}: {turno.carga}")

class Empleado:
    def __init__(self, id, nombre, telefono, direccion):
        self.id = id
        self.nombre = nombre
        self.telefono = telefono
        self.direccion = direccion

class Camion:
    def __init__(self, id, ruta, conductor, asistentes):
        self.id = id
        self.ruta = ruta
        self.conductor = conductor
        self.asistentes = asistentes
        self.turno = None

class Ruta:
    def __init__(self):
        self.puntosGeograficos = []

    def Iterar(self):
        for punto in self.puntosGeograficos:
            print(f"Latitud: {punto.latitud}, Longitud: {punto.longitud}, Dirección: {punto.direccion}")

class PuntoGeografico:
    def __init__(self, direccion, latitud, longitud):
        self.direccion = direccion
        self.latitud = latitud
        self.longitud = longitud

class Turno:
    def __init__(self, id, camion, inicio, fin, carga):
        self.id = id
        self.camion = camion
        self.inicio = inicio
        self.fin = fin
        self.carga = carga

```

```python
# Creación de objetos TrashCity, Empleado, Ruta, PuntoGeografico y Camion
trash_city = TrashCity()

empleado1 = Empleado("1", "Juan", "123456789", "Calle A")
empleado2 = Empleado("2", "María", "987654321", "Calle B")
trash_city.contratarEmpleado(empleado1)
trash_city.contratarEmpleado(empleado2)

ruta = Ruta()
punto1 = PuntoGeografico("Dirección 1", 40.123, -75.456)
punto2 = PuntoGeografico("Dirección 2", 40.456, -75.789)
ruta.puntosGeograficos.append(punto1)
ruta.puntosGeograficos.append(punto2)

camion = Camion("1", ruta, empleado1, [empleado2])
trash_city.agregarCamion(camion)

# Creación de un turno con carga
inicio = datetime(2023, 5, 19, 9, 0)
fin = datetime(2023, 5, 19, 12, 0)
carga = {"vidrio": 2.5, "papel": 1.0, "plastico": 3.0, "metal": 0.5, "organicos": 2.0}
turno = Turno("T1", camion, inicio, fin, carga)
trash_city.turnos.append(turno)

# Obtener la cantidad de vidrio en un turno de una fecha específica
fecha_especifica = datetime(2023, 5, 19)
trash_city.cargaDiaN(fecha_especifica)
```
